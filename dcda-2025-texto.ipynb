```python
import pandas as pd 

from transformers import pipeline
```

```python
train = pd.read_csv('../input/petfinder-adoption-prediction/train/train.csv').set_index("PetID")
test = pd.read_csv('../input/petfinder-adoption-prediction/test/test.csv').set_index("PetID")
```

```python
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True, device=0)

#classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)


descr = classifier(train.Description.fillna("empty").astype(str).values.tolist(), max_length=512, truncation=True)
descr = pd.DataFrame([{r["label"]: r["score"] for r in res} for res in descr], index=train.index)
descr
```

```python
descr_test = classifier(test.Description.fillna("empty").astype(str).values.tolist(), max_length=512, truncation=True)
descr_test = pd.DataFrame([{r["label"]: r["score"] for r in res} for res in descr_test], index=test.index)
descr_test
```

```python
descr.to_parquet("train_text.parquet")
descr_test.to_parquet("test_text.parquet")
```
